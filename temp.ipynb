```python
#!/usr/bin/env python
# coding: utf-8

import os
import requests

# Retrieve your API token from the environment (same as in your shell).
# Make sure you have exported DEEPINFRA_TOKEN in your shell environment before running this.
# e.g., export DEEPINFRA_TOKEN="your_actual_token"
token = os.environ.get("DEEPINFRA_TOKEN")

url = "https://api.deepinfra.com/v1/openai/chat/completions"
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {token}"
}

data = {
    "model": "deepseek-ai/DeepSeek-R1",
    "messages": [
        {
            "role": "user",
            "content": "Write a short story broken into parts about an attorney who gets the nomination for attorney general."
        }
    ]
}

response = requests.post(url, headers=headers, json=data)

print(response.json())
```
